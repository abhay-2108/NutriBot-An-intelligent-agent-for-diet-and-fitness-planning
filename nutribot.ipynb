```python
import os

folder = 'Fitabase Data 4.12.16-5.12.16'
csv_files = [f for f in os.listdir(folder) if f.endswith('.csv')]

for idx, file in enumerate(csv_files, start=1):
    globals()[f'df{idx}'] = pd.read_csv(os.path.join(folder, file))
```

```python
# -- useful data : df1, df2, df3, df4, df5,df17,df18  
df1.head()  # Daily Activity data
```

```python
df1 = df1[['TotalSteps', 'TotalDistance']]
df1.head()
```

```python
scaler = StandardScaler()
df1_scaled = scaler.fit_transform(df1)
```

```python
df1_scaled = pd.DataFrame(df1_scaled, columns=df1.columns, index=df1.index)
df1_scaled.head()
```

```python
kmeans1 = KMeans(n_clusters=3, random_state=42)
clusters = kmeans1.fit_predict(df1_scaled)
df1_scaled['Cluster'] = clusters
df1_scaled.head()
```

```python
import joblib
import os

os.makedirs("model", exist_ok=True)

joblib.dump(kmeans1, "model/daily-df1.pkl") 
joblib.dump(scaler, "model/fitness_scaler.pkl")
```
